```python
%%writefile app.py
import streamlit as st
from fpdf import FPDF

st.title("Simple AI Resume Builder")

# --- User Input ---
name = st.text_input("Full Name")
email = st.text_input("Email")
phone = st.text_input("Phone")
education = st.text_area("Education")
experience = st.text_area("Work Experience")
skills = st.text_area("Skills")

resume_type = st.selectbox("Choose Resume Type", ["Classic", "Modern", "Professional"])

# --- Generate Resume ---
if st.button("Generate Resume"):
    if resume_type == "Classic":
        resume_text = f"""
{name}
Email: {email} | Phone: {phone}

Education:
{education}

Work Experience:
{experience}

Skills:
{skills}
"""
    elif resume_type == "Modern":
        resume_text = f"""
=== {name.upper()} ===
ðŸ“§ {email}   ðŸ“ž {phone}

--- Education ---
{education}

--- Work Experience ---
{experience}

--- Skills ---
{skills}
"""
    elif resume_type == "Professional":
        resume_text = f"""
Name: {name}
Contact Info: {email} | {phone}

Education:
{education}

Professional Experience:
{experience}

Key Skills:
{skills}
"""
    st.text_area("Your Resume", resume_text, height=300)

    # --- Generate PDF ---
    pdf = FPDF()
    pdf.add_page()
    pdf.set_font("Arial", size=12)
    for line in resume_text.split("\n"):
        pdf.multi_cell(0, 10, line)
    pdf.output("resume.pdf")

    with open("resume.pdf", "rb") as f:
        st.download_button("Download PDF", f, file_name="resume.pdf")
```
